```python
def calculate_iou(box1, box2):
    """
    计算两个目标框之间的IoU。
    
    :param box1: 第一个目标框的坐标，格式为 (x1, y1, x2, y2)
    :param box2: 第二个目标框的坐标，格式为 (x1, y1, x2, y2)
    :return: 两个目标框之间的IoU
    """

    # 获取第一个目标框的坐标，_后的数值表示是第几个框
    x1_1, y1_1, x2_1, y2_1 = box1
    
    # 获取第二个目标框的坐标
    x1_2, y1_2, x2_2, y2_2 = box2
    
    # 计算交集区域的坐标
    xi1 = max(x1_1, x1_2)  # 表示两个框中最左上角的x最大
    yi1 = max(y1_1, y1_2)  # 表示两个框中最左上角的y最大
    xi2 = min(x2_1, x2_2)  # 表示两个框中最右下角的x最小
    yi2 = min(y2_1, y2_2)  # 表示两个框中最右下角的y最小
    
    # 计算交集区域的宽度和高度
    inter_width = max(xi2 - xi1, 0)
    inter_height = max(yi2 - yi1, 0)
    
    # 计算交集区域的面积
    inter_area = inter_width * inter_height
    
    # 计算两个目标框的面积
    box1_area = (x2_1 - x1_1) * (y2_1 - y1_1)
    box2_area = (x2_2 - x1_2) * (y2_2 - y1_2)
    
    # 计算并集区域的面积
    union_area = box1_area + box2_area - inter_area
    
    # 计算IoU
    iou = inter_area / union_area
    
    return iou

# 示例使用
box1 = (50, 50, 150, 150)
box2 = (100, 100, 200, 200)

iou = calculate_iou(box1, box2)
print(f"两个目标框之间的IoU为: {iou:.4f}")
```

```python
import os
import numpy as np
from PIL import Image
import nibabel as nib

# PNG 图像文件夹路径
png_folder = "/mnt/hdd1/zhulu/blood_stage2/blood_anno/PNG/Patient-0018470179/Series-Ax SWAN new"
# 要保存的 NIfTI 文件路径
nii_output_path = "/mnt/hdd1/zhulu/blood_stage2/blood_anno/PNG/Patient-0018470179/data.nii"

# 获取 PNG 文件列表并排序（确保按照正确的顺序加载）
png_files = sorted([f for f in os.listdir(png_folder) if f.endswith('.png')])

# 初始化一个列表来存储每个图层
slices = []

# 读取每个 PNG 文件并将其添加到列表中
for file in png_files:
    img_path = os.path.join(png_folder, file)
    # 使用 Pillow 读取图像
    img = Image.open(img_path)
    # 将图像转换为 numpy 数组，并将其添加到 slices 列表中
    slices.append(np.array(img))

# 将列表转换为 3D numpy 数组（维度：深度, 高度, 宽度）
image_3d = np.stack(slices, axis=-1)

# 创建 NIfTI 图像
nii_image = nib.Nifti1Image(image_3d, np.eye(4))

# 保存为 NIfTI 文件
nib.save(nii_image, nii_output_path)

print(f"NIfTI 文件已保存到: {nii_output_path}")
```

```python
import os 
#此处训练集和验证集都要进行删除
#训练集如果不删除其他标签数据报错明显，但是测试集不删除报错提示不明显
""" 注意是train和val都要删除 """
# train_txt="/mnt/hdd1/zhulu/hospital/labels/train"  
# txt_files=os.listdir(train_txt)

def adjust_yolo_file(file_dir):
    txt_files=os.listdir(file_dir)
    for file in txt_files:
        file_path=os.path.join(file_dir,file)
        with open(file_path, 'r') as file:
            lines = file.readlines()
        
        # 选出符合出血点标签的行
        updated_lines = [line for line in lines if line.strip().split()[0] == "0"]
        
        if len(updated_lines)==0: #如果文件没有出血点的标注，则直接删除文件
            os.remove(file_path)
        else:
            # 将更新后的内容写回文件
            with open(file_path, 'w') as f:
                f.writelines(updated_lines)
        
```

```python
import cv2
import numpy as np

v_path="/mnt/hdd1/zhulu/longvideo_grid/videos/238.mp4"
# 打开视频文件
cap = cv2.VideoCapture(v_path)
frame_list=[]
key_frame_list=[]
# 检查视频是否成功打开
if not cap.isOpened():
    print("Error: Could not open video.")
else:
    while True:
        ret, frame = cap.read()
        
        if not ret:
            break
        # 显示每一帧
        frame_list.append(frame)
        
    # 释放视频捕获对象
    cap.release()
    cv2.destroyAllWindows()
    
# 确保 frame_list 至少有两帧
if len(frame_list) >= 2:
    # 获取第一帧和第二帧
    frame1 = frame_list[0]
    frame2 = frame_list[100]

    # 计算像素值绝对差
    abs_diff = cv2.absdiff(frame1, frame2)
    
    # 显示或保存结果
    cv2.imwrite("temp/abs_diff.png", abs_diff)
    print("Absolute difference calculated and saved as 'temp/abs_diff.png'")
else:
    print("Not enough frames in frame_list to calculate absolute difference")

key_frame_list.append(frame_list[0])
avg_diff_list=[]
for i in range(len(frame_list)-1):
    avg_diff=np.sum(cv2.absdiff(frame_list[i], frame_list[i+1]))/(frame_list[i].shape[0]*frame_list[i].shape[1]*frame_list[i].shape[2])    
    avg_diff_list.append(avg_diff)
    
```

```python
len(avg_diff_list),len(frame_list)
top_10_avg_diff = sorted(avg_diff_list, reverse=True)[:48]
top_10_indices = [avg_diff_list.index(diff) for diff in top_10_avg_diff]
top_10_indices
sorted(top_10_indices)
for indice in top_10_indices:
    key_frame_list.append(frame_list[indice+1])
    
```

```python
import numpy as np

# 假设所有帧的大小相同
frame_height, frame_width, channels = key_frame_list[0].shape

# 计算拼接后的图像大小
rows = (len(key_frame_list) + 2) // 3  # 每行3个图像
cols = 3
output_height = rows * frame_height
output_width = cols * frame_width

# 创建一个空的数组来存储拼接后的图像
output_image = np.zeros((output_height, output_width, channels), dtype=key_frame_list[0].dtype)

# 将帧按顺序拼接到输出图像中
for idx, frame in enumerate(key_frame_list):
    row = idx // cols
    col = idx % cols
    output_image[row * frame_height:(row + 1) * frame_height, col * frame_width:(col + 1) * frame_width, :] = frame

# 保存拼接后的图像
cv2.imwrite("temp/stitched_image.png", output_image)
print("Stitched image saved as 'temp/stitched_image.png'")
```

### 形态学筛选

```python
import cv2
import numpy as np
def find_and_update(matrix, start_point, threshold=2):
    height, width = matrix.shape
    visited = np.zeros_like(matrix, dtype=bool)  # To track visited points
    result_matrix = np.full_like(matrix, 255, dtype=np.uint8)  # Initialize with 255
    queue = [start_point]  # Start from the given point
    visited[start_point] = True

    # Get the seed point value
    seed_value = matrix[start_point]

    # Directions for 4-neighborhood
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    while queue:
        x, y = queue.pop(0)
        result_matrix[x, y] = 0  # Set companion points to 0

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if 0 <= nx < height and 0 <= ny < width and not visited[nx, ny]:
                # Compare with the seed point value
                if abs(int(matrix[nx, ny]) - int(seed_value)) <= threshold:
                    visited[nx, ny] = True
                    queue.append((nx, ny))

    return result_matrix

def find_min_in_center(matrix, margin=5):
    height, width = matrix.shape
    # Restrict the search area to avoid the edges
    restricted_area = matrix[margin:height-margin, margin:width-margin]
    # Find the minimum value and position within the restricted area
    min_value = np.min(restricted_area)
    min_position = np.unravel_index(np.argmin(restricted_area), restricted_area.shape)
    # Adjust position to match the original matrix indices
    min_position = (min_position[0] + margin, min_position[1] + margin)
    return min_value, min_position


def calculate_circularity(binary_image):
    # Find contours
    binary_image = cv2.bitwise_not(binary_image)
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Initialize variables to store the largest contour properties
    largest_contour = None
    largest_area = 0
    if len(contours) <= 3 and len(contours[0]) <= 3:
        return 1
    
    # Iterate through contours to find the largest one
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > largest_area:
            largest_area = area
            largest_contour = contour
    
    # Calculate circularity if a contour is found
    circularity = None
    if largest_contour is not None:
        perimeter = cv2.arcLength(largest_contour, True)
        if perimeter > 0:  # Avoid division by zero
            circularity = (4 * np.pi * largest_area) / (perimeter ** 2)
    
    return circularity
```

```python
from pathlib import Path

def rename_folders(directory):
    # 将目录路径转换为 Path 对象
    path = Path(directory)
    # 遍历目录下的所有文件夹
    for folder in path.iterdir():
        if folder.is_dir() and not folder.name.startswith('Patient'):
            # 新文件夹名
            new_folder_name = f"Patient-{folder.name}"
            new_folder_path = folder.with_name(new_folder_name)
            
            # 重命名文件夹
            folder.rename(new_folder_path)
            print(f'Renamed: {folder.name} -> {new_folder_name}')

# 指定要操作的目录
directory = '/mnt/hdd1/zhulu/blood_stage2/total'
rename_folders(directory)
```
